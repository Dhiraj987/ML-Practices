A sequence that arises in ecology as a model for population growth is defined by the logistic difference equation Pn+1 = kpn(1 - Pn) where Pn measures the size of the population of the n-th generation of a single species. To keep the numbers manageable, Pn is a fraction of the maximal size of the population, so its 0 to 1

```python
import matplotlib.pyplot as plt
```

# for one data

```python
p1=[0.5]
k1=2.8
```

```python
for i in range(30):
    p1next=k1*p1[-1]*(1-p1[-1])
    p1.append(p1next)
```

```python
plt.plot(p1)
```

# for two data

```python
# 'p1' with 'k1'
p1 = [0.5]
# 'p2' with 'k2'
p2 = [0.5]
k1 = 3.7
k2 = 3.9
```

```python
# Calulating 30 terms for 'p1' and 'p2'
for i in range(10):
   P1next = k1*p1[-1]*(1-p1[-1])
   p1.append(P1next)
   P2next = k2*p2[-1]*(1-p2[-1])
   p2.append(P2next)
```

```python
# Plotting graph for 30 terms
plt.plot(p1)
plt.plot(p2)
plt.xlabel('P values')
# Setting legend in graph
plt.legend(["k = "+str(k1), "k = "+str(k2)])
plt.show()
```
