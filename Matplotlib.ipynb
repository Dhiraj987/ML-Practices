```python
import  matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('TITLE')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#OO
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('TITLE')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Outside')
axes1.set_xlabel('X LABEL')

axes2.plot(y,x)
axes2.set_title('Inside')
axes2.set_xlabel('X LABEL')
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('THE TITLW')
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig,axes= plt.subplots(2,1,figsize=(8,2))

axes[0].plot(x,y)
plt.tight_layout()
```

```python
fig.savefig('first.pdf',dpi=200)
```

```python
fig=plt.figure()
axx=fig.add_axes([0,0,1,1])
axx.plot(x,x**2,label='x Sq')
axx.plot(x,x**3,label='x C')

axx.legend(loc=10)
```
